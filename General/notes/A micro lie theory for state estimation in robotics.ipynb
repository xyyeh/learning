### Lie group and lie algebra

1. The Lie algebra $T_\epsilon M$ (red plane) is the tangent space to the Lie group's manifold M (here represented as the blue sphere) at the identity $\epsilon$.
2. Through the exponential map, each straight path $\mathbf{v} t$ through the origin on the Lie algebra produces a path $exp(\mathbf{v} t)$ around the manifold which runs along the respective geodesic.
3. The Lie algebra of the Lie group is thus the tangent space at the identity. Vectors of a particular tangent space can be transformed to the Lie algebra through an **adjoint** transformation.
4. The structure of the Lie algebra can be exposed by time differentiation of the group constraint (usually the inverse relationship). For example, the constraints for an $SO(3)$ structure is such that $R^T R = R R^T = I$. Taking the time derivative yields $\dot R^T R = R^T \dot R$. This implies that $\dot R^T R$ is skew-symmetric. At the identity, $R = I$, the Lie algebra of $SO(3)$, $\mathfrak{so}(3)$ is $\dot R = [\omega]_\times$.

#### Mappings between the manifold and the Lie algebra

To simplify the notation, the following holds:
\begin{align}
\mathit{X} &= Exp(\tau) \triangleq exp(\tau^\wedge) \\
\tau &= Log(\mathit{X}) \triangleq log(\mathit{X})^\vee
\end{align}
where $\wedge: \mathbb{R}^3 \mapsto \mathfrak{so}(3)$ and $\vee: \mathfrak{so}(3) \mapsto \mathbb{R}^3$

#### Plus and minus operations on manifolds
1. The $\oplus$ and $\ominus$ operators allow us to introduce increments between elements of a manifold and express them in the tangent vector space. They combine Exp/Log operations with one composition.
2.  $\epsilon$ is the identity. Since the operations are non-commutative, we define the left and right versions:
\begin{align}
right-\oplus: \mathit{Y} = \mathit{X} \oplus ~^\mathit{X}\tau \triangleq \mathit{X} \circ Exp(^\mathit{X}\tau) \in \mathit{M} \\
right-\ominus: ~^\mathit{X} \tau = \mathit{Y} \ominus \mathit{X} \triangleq Log(\mathit{X}^{-1} \circ \mathit{Y}) \in T_{\mathit{X}}M
\end{align}

\begin{align}
left-\oplus: \mathit{Y} = ~^\mathit{\epsilon}\tau \oplus \mathit{X} \triangleq Exp(^\mathit{\epsilon}\tau) \circ \mathit{X}  \in \mathit{M} \\
left-\ominus: ~^\mathit{\epsilon} \tau = \mathit{Y} \ominus \mathit{X} \triangleq Log(\mathit{Y} \circ \mathit{X}^{-1}) \in T_{\mathit{\epsilon}}M
\end{align}

#### Adjoint operations
1. The adjoint $\mathbf{Ad}_{\mathit{X}}$ maps tangent vector $^\mathit{X}\tau$ to $^\mathit{\epsilon}\tau$. Since $^\mathit{\epsilon}\tau \oplus \mathit{X} = \mathit{X} \oplus ~^\mathit{\mathit{X}}\tau$, the relation between the local and global tangent elements can be derived to be $^\mathit{\epsilon}\tau^\wedge = \mathit{X} ~^\mathit{X} \tau^\wedge \mathit{X}^{-1}$
2. The adjoint on $SE(3)$ is given by $Ad_M = \begin{bmatrix} R & [t]_\times R \\ 0 & R \\ \end{bmatrix}$ where $M = \begin{bmatrix} R & t \\ 0 & 1 \end{bmatrix}$ and $~^M \tau = \begin{bmatrix} \rho \\ \theta \end{bmatrix}$. The adjoint matrix can be derived from $Ad_M \tau = (M ~^M \tau^\wedge M^{-1})^\vee$
